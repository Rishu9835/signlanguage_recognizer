```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

import os
import PIL
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping


```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import os

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2]
)

input_dir = 'sign_dataset'   # Folder with cropped A to Y images
output_dir = 'dataset'

os.makedirs(output_dir, exist_ok=True)

for letter in os.listdir(input_dir):
    img_path = os.path.join(input_dir, letter)
    img = load_img(img_path)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    
    save_dir = os.path.join(output_dir, letter)
    os.makedirs(save_dir, exist_ok=True)
    
    i = 0
    for batch in datagen.flow(x, batch_size=1, save_to_dir=save_dir, save_prefix=letter, save_format='jpg'):
        i += 1
        if i > 50:  # generate 50 augmented images per letter
            break
```

```python
sign_path={
    'A':list(data_dir.glob('A.jpg/*')),
    'B':list(data_dir.glob('B.jpg/*')),
    'C':list(data_dir.glob('C.jpg/*')),
    'D':list(data_dir.glob('D.jpg/*')),
    'E':list(data_dir.glob('E.jpg/*')),
    'F':list(data_dir.glob('F.jpg/*')),
    'G':list(data_dir.glob('G.jpg/*')),
    'H':list(data_dir.glob('H.jpg/*')),
    'I':list(data_dir.glob('I.jpg/*')),
    'K':list(data_dir.glob('K.jpg/*')),
    'L':list(data_dir.glob('L.jpg/*')),
    'M':list(data_dir.glob('M.jpg/*')),
    'N':list(data_dir.glob('N.jpg/*')),
    'O':list(data_dir.glob('O.jpg/*')),
    'P':list(data_dir.glob('P.jpg/*')),
    'Q':list(data_dir.glob('Q.jpg/*')),
    'R':list(data_dir.glob('R.jpg/*')),
    'S':list(data_dir.glob('S.jpg/*')),
    'T':list(data_dir.glob('T.jpg/*')),
    'U':list(data_dir.glob('U.jpg/*')),
    'V':list(data_dir.glob('V.jpg/*')),
    'W':list(data_dir.glob('W.jpg/*')),
    'X':list(data_dir.glob('X.jpg/*')),
    'Y':list(data_dir.glob('Y.jpg/*')),
}
```

```python
sign_labels={
    'A':0,
    'B':1,
    'C':2,
    'D':3,
    'E':4,
    'F':5,
    'G':6,
    'H':7,
    'I':8,
    'K':9,
    'L':10,
    'M':11,
    'N':12,
    'O':13,
    'P':14,
    'Q':15,
    'R':16,
    'S':17,
    'T':18,
    'U':19,
    'V':20,
    'W':21,
    'X':22,
    'Y':23,
}
```

```python
X,y =[],[]
for name,images in sign_path.items():
    for image in images:
        img=cv2.imread(str(image))
        resized_image=cv2.resize(img,(180,180))
        X.append(resized_image)
        y.append(sign_labels[name])
```

```python
X=np.array(X)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train_scaled = X_train/255
X_test_scaled = X_test/255
```

```python
num_classes=24
cnn = Sequential([
    #cnn
    layers.Conv2D(16 ,3 , padding='same' , activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32 ,3 , padding='same' , activation='relu'),
    layers.MaxPooling2D(),
    # layers.Conv2D(64 ,3 , padding='same' , activation='relu'),
    # layers.MaxPooling2D(),

    #dense network
    layers.Flatten(),
    layers.Dense(50,activation='relu'),
    layers.Dense(num_classes)
])
cnn.compile(optimizer='adam',
            loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
            metrics=['accuracy']
            
             )
cnn.fit(X_train_scaled,y_train,epochs=25)
```

```python
cnn.evaluate(X_test_scaled,y_test)
```
